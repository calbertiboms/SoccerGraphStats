```python
from statsbombpy import sb
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
compid = 11
seasonid= 90
tournament = sb.matches(competition_id=compid,season_id=seasonid)
tournament
```

```python
  barca_home = tournament[tournament.home_team == 'Barcelona']
  barca_away = tournament[tournament.away_team == 'Barcelona']
  barca = pd.concat([barca_home,barca_away])
```

```python
barca
```

```python
barcaplayers = []
for idmatch in barca.match_id.unique():
  soccer_match = sb.events(match_id = idmatch)
  soccer_match=soccer_match[soccer_match.team == 'Barcelona']
  for player in soccer_match.player.unique():
    if player not in barcaplayers:
      barcaplayers.append(player)
```

```python
from urllib.request import urlopen
from PIL import Image, ImageDraw
from mplsoccer import Pitch
import numpy as np
import matplotlib.pyplot as plt

def draw_player(image_url, x, y, ax):

    image_player = Image.open(urlopen(image_url))


    width, height = image_player.size
    radius = int(min(width, height) / 2)
    center = width // 2, height // 2.5
    mask = Image.new('L', (width, height), 0)
    draw = ImageDraw.Draw(mask)
    draw.ellipse((center[0]-radius, center[1]-radius, center[0]+radius, center[1]+radius), fill=255)
    circular_image = Image.composite(image_player, Image.new('RGBA', image_player.size, (0, 0, 0, 0)), mask)
    circular_image = circular_image.crop((center[0]-radius, center[1]-radius, center[0]+radius, center[1]+radius))

    # Convert to a format suitable for mplsoccer
    circular_image_np = np.array(circular_image)



    # Adjust the size as per your requirements. Here, I'm using height=10 units.
    ax_image = pitch.inset_image(x, y, circular_image_np, height=10, ax=ax)





```
