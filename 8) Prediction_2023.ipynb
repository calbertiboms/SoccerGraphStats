```python
n = 3

# Function to create shifted columns for a dataframe
def create_shifted_columns_for_team(df, team_col, n, prefix):
    team_dfs = []
    for team in df[team_col].unique():
        team_df = df[df[team_col] == team].copy()
        for col in ['density', 'transitivity', 'average_clustering', 'diameter']:
            for i in range(1, n + 1):
                shifted_col_name = f"{prefix}_{col}_{i}"
                team_df[shifted_col_name] = team_df[f"{prefix}_{col}"].shift(i)
        team_dfs.append(team_df)
    return pd.concat(team_dfs)

# Apply the function to create shifted columns for home and away teams
home_df_shifted = create_shifted_columns_for_team(merged_df, 'home_team', n, 'home')
away_df_shifted = create_shifted_columns_for_team(merged_df, 'away_team', n, 'away')

# Select only the shifted columns along with match_id for merging
home_cols_shifted = ['match_id'] + [col for col in home_df_shifted.columns if 'home_' in col and col not in merged_df.columns]
away_cols_shifted = ['match_id'] + [col for col in away_df_shifted.columns if 'away_' in col and col not in merged_df.columns]

# Merge the original dataframe with the shifted data for the home team
merged_home_df = pd.merge(merged_df, home_df_shifted[home_cols_shifted], on="match_id") # on is used to merge based on match_id

# Merge the result with the shifted data for the away team
final_merged_df = pd.merge(merged_home_df, away_df_shifted[away_cols_shifted], on="match_id")

# Define the desired columns in the correct order
columns = [
    'match_id', 'home_team', 'away_team', 'outcome',
    'home_density', 'home_diameter', 'home_transitivity', 'home_average_clustering',
    'away_density', 'away_diameter', 'away_transitivity', 'away_average_clustering',
    'home_score', 'away_score'
] + home_cols_shifted[1:] + away_cols_shifted[1:]  # Exclude 'match_id' from shifted columns

final_df = final_merged_df[columns]
final_df.tail()
```

```python
average_density = (merged_df['home_density'].mean() + merged_df['away_density'].mean()) / 2
average_diameter = (merged_df['home_diameter'].mean() + merged_df['away_diameter'].mean()) / 2
average_transitivity = (merged_df['home_transitivity'].mean() + merged_df['away_transitivity'].mean()) / 2
average_clustering = (merged_df['home_average_clustering'].mean() + merged_df['away_average_clustering'].mean()) / 2
```

```python


for i in range(1, n + 1):
    final_df[f'home_density_{i}'].fillna(average_density, inplace=True) #inplace true means that I am modifying the df, not creating a new one
    final_df[f'away_density_{i}'].fillna(average_density, inplace=True)
    final_df[f'home_diameter_{i}'].fillna(average_diameter, inplace=True)
    final_df[f'away_diameter_{i}'].fillna(average_diameter, inplace=True)
    final_df[f'home_transitivity_{i}'].fillna(average_transitivity, inplace=True)
    final_df[f'away_transitivity_{i}'].fillna(average_transitivity, inplace=True)
    final_df[f'home_average_clustering_{i}'].fillna(average_clustering, inplace=True)
    final_df[f'away_average_clustering_{i}'].fillna(average_clustering, inplace=True)

final_df.head()
```

```python
# Function to search for the history
def history_team(matches_df, selected_team, idmatch):
    # Find the index for the row that matches the given idmatch
    match_index = matches_df.index[matches_df['match_id'] == idmatch].tolist()



    # We have a unique match, so we get its index
    up_to_index = match_index[0]-1


    # Initialize dictionaries to store the histories
    winning_history = {}
    losing_history = {}
    ties_history = {}
    goals_history = {}
    goals_avg_history = {}


    # Filter the matches DataFrame to only include matches before the specified index
    tournament = matches_df.loc[:up_to_index]
    team_home = tournament[tournament.home_team == selected_team]
    team_away = tournament[tournament.away_team == selected_team]
    total_matches = pd.concat([team_home,team_away])

    wins = 0
    loses = 0
    ties = 0
    total_goals = 0

    # Iterate through home matches
    for index in team_home.index:
        # Initialize variables for this scope
        winner = None
        loser = None

        match_row = tournament.loc[index]
        home_team = match_row['home_team']
        away_team = match_row['away_team']
        home_score = match_row['home_score']
        away_score = match_row['away_score']


        if home_score > away_score:
            winner = home_team
            loser = away_team
        elif home_score < away_score:
            winner = away_team
            loser = home_team


        if winner == selected_team:
            wins += 1
        elif loser == selected_team:
            loses += 1
        else:
            ties += 1
        total_goals += home_score

# We see the df where the team is the away team
    for idmatch in team_away.match_id.unique():

      winner = None
      loser = None
      match_row = tournament[tournament.match_id == idmatch]
      home_team = match_row.iloc[0]['home_team']
      away_team = match_row.iloc[0]['away_team']
      if match_row.home_score.iloc[0] != match_row.away_score.iloc[0]:
        winner = home_team if match_row.home_score.iloc[0] > match_row.away_score.iloc[0] else away_team
        loser = home_team if match_row.home_score.iloc[0] < match_row.away_score.iloc[0] else away_team
      if winner == selected_team:
        wins += 1
      elif loser == selected_team:
        loses += 1
      else:
        ties += 1
      total_goals += match_row.away_score.iloc[0]

    # Compute histories
    if len(total_matches)==0:
      winning_history[selected_team] = None
      losing_history[selected_team] = None
      ties_history[selected_team] = None
      goals_history[selected_team]=None
      goals_avg_history[selected_team]=None
    else:
      winning_history[selected_team] = (wins/len(total_matches))
      losing_history[selected_team] = (loses/len(total_matches))
      ties_history[selected_team] = (ties/len(total_matches))
      goals_history[selected_team]=(total_goals)
      goals_avg_history[selected_team]=(total_goals/len(total_matches))


    return winning_history, losing_history, ties_history, goals_history, goals_avg_history
```

```python
matches_history = []
```

```python
for index, row in final_df.iterrows():
    # Call the history_team function for both the home and away team for each match
    winning_history_home = None


    winning_history_home, losing_history_home, ties_history_home, goals_history_home, goals_avg_history_home = history_team(
       final_df, row['home_team'], row['match_id'])

    winning_history_away, losing_history_away, ties_history_away, goals_history_away, goals_avg_history_away = history_team(
        final_df, row['away_team'], row['match_id'])

    # Extract the relevant data and append to the list
    match_data = {
        'match_id': row['match_id'],
        'winning_history_home': winning_history_home[row['home_team']],
        'winning_history_away': winning_history_away[row['away_team']],
        'losing_history_home': losing_history_home[row['home_team']],
        'losing_history_away': losing_history_away[row['away_team']],
        'tie_history_home': ties_history_home[row['home_team']],
        'tie_history_away': ties_history_away[row['away_team']],
        'avg_goals_home': goals_avg_history_home[row['home_team']],
        'avg_goals_away': goals_avg_history_away[row['away_team']],
    }
    matches_history.append(match_data)

# Create a DataFrame from the collected history data
history_df = pd.DataFrame(matches_history)
```

# History of the pair of teams

```python
# This funciton calculates the historic stats of the outcomes of the matches between the 2 teams at stake in a certain match
# It uses all the previous matches
def calculate_head_to_head_stats(matches_df, idmatch):
  teams = matches_df[matches_df.match_id == idmatch]
  team1 = teams.home_team.iloc[0]
  team2 = teams.away_team.iloc[0]
  team1_wins = 0
  team1_losses = 0
  ties = 0
  total_matches = 0
  match_index = matches_df.index[matches_df['match_id'] == idmatch].tolist()
  # We want to use previous data only
  up_to_index = match_index[0]-1
  tournament = matches_df.loc[:up_to_index]
  # Loop through each match

  for index, row in tournament.iterrows():
        # Check if the current match is between team1 and team2
        if ((row['home_team'] == team1 and row['away_team'] == team2) or
            (row['home_team'] == team2 and row['away_team'] == team1)):
            total_matches += 1

            # Calculate outcomes
            if row['home_team'] == team1:
                if row['home_score'] > row['away_score']:
                    team1_wins += 1
                elif row['home_score'] < row['away_score']:
                    team1_losses += 1
                else:
                    ties += 1
            else:
                if row['away_score'] > row['home_score']:
                    team1_wins += 1
                elif row['away_score'] < row['home_score']:
                    team1_losses += 1
                else:
                    ties += 1

    # Calculate percentages
  win_percentage = (team1_wins / total_matches)  if total_matches > 0 else None
  loss_percentage = (team1_losses / total_matches) if total_matches > 0 else None
  tie_percentage = (ties / total_matches) if total_matches > 0 else None

  return {
        'match_id': idmatch,
        'home_team': team1,
        'away_team': team2,
        'pair_win_home': win_percentage,
        'pair_win_away': loss_percentage,
        'pair_tie': tie_percentage}
```

```python
pair_df_rows = []
for idmatch in final_df['match_id'].unique():
    pair_stats = calculate_head_to_head_stats(final_df, idmatch)
    pair_df_rows.append(pair_stats)

# Convert the list of dictionaries to a DataFrame
pair_df = pd.DataFrame(pair_df_rows)
```

```python
# We handle NAN values by replacing them for the default probabilities, 33%
columns_to_replace = pair_df.columns
pair_df[columns_to_replace] = pair_df[columns_to_replace].fillna(0.33)
```

# Machine Learning Model

```python
X_train = df_train.drop('outcome', axis=1)
y_train = df_train['outcome']
X_test = df_predict.drop('outcome', axis=1)
y_test = df_predict['outcome']
```

```python
from lightgbm import LGBMClassifier
from sklearn.metrics import classification_report, accuracy_score

lgbm = LGBMClassifier()

lgbm.fit(X_train, y_train)

y_pred = lgbm.predict(X_test)

# Evaluate model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Confusion data matrix
conf_matrix_data = confusion_matrix(y_test, y_pred)
labels = sorted(list(set(y_test)))

# Create a heatmap of the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_data, annot=True, fmt='d', cmap='RdPu', xticklabels=labels, yticklabels=labels)

plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')

plt.show()
```
